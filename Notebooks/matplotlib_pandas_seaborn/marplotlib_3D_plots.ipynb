```python
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib: 3D Ploting 

3D plotting is particularly useful when visualizing complex datasets or understanding relationships between three variables. 

A three-dimensional axes can be created by passing the keyword `projection='3d'` to any of the normal axes creation routines.

Note that prior to Matplotlib 3.2.0, it was necessary to explicitly import the mpl_toolkits.mplot3d module to make the '3d' projection to Figure.add_subplot.

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

## Basic 3D plot

Let's begin with a basic 3D plot. We'll plot a line connecting three points in 3D space:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define data
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

# Plot the line
ax.plot(x, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic 3D Plot')

# Show the plot
plt.show()
```

## Scatter plot in 3D

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate random data
np.random.seed(0)
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# Plot the scatter plot
ax.scatter(x, y, z, c='r', marker='o')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Scatter Plot in 3D')

# Show the plot
plt.show()
```

## Surface plot

A surface plot represents a 3D surface or a function in three dimensions. Let's create a surface plot using the famous [Rosenbrock function](https://en.wikipedia.org/wiki/Rosenbrock_function).

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate data for the surface plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot')

# Show the plot
plt.show()
```

## Wireframe plot

A wireframe plot visualizes a surface plot with only its wireframe representation. Let's create a wireframe plot of the same Rosenbrock function.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the wireframe
ax.plot_wireframe(X, Y, Z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Wireframe Plot')

# Show the plot
plt.show()
```

## Contour plot

A contour plot represents the level curves of a 3D surface in two dimensions. Let's create a contour plot of the Rosenbrock function.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the contour
ax.contour(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Contour Plot')

# Show the plot
plt.show()
```
