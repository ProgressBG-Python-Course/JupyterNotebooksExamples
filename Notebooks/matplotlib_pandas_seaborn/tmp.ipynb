```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read in the drinks.csv file
drinks = pd.read_csv('../../datasets/various/drinks.csv', nrows=5).drop(columns=['spirit_servings','total_litres_of_pure_alcohol'])
drinks
```

```python
# Create a 3D scatter plot of total liters of pure alcohol consumed per capita 
# by continent and country
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')


# continent_map = {s: i for i, s in enumerate(set(drinks.continent))}
# numeric_strings = [string_map[s] for s in drinks.continent]

continent_map = {s: i for i, s in enumerate(set(drinks.continent))}
country_map = {s: i for i, s in enumerate(set(drinks.country))}

x = drinks.continent.map(continent_map)
y = drinks.beer_servings
z = drinks.wine_servings

print(continent_map)
print(x)
print(y)

xticks = continent_map.values()
xtickslabels = continent_map.keys()

ax.set(xticks=list(xticks), xticklabels=list(xtickslabels)) 

ax.scatter(x,y,z)
ax.set_xlabel('Continent')
ax.set_ylabel('Beer Servings')
ax.set_zlabel('Wine Servings')

plt.show()
```
