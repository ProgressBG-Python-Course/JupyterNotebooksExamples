```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read in the drinks.csv file
drinks = pd.read_csv('../../datasets/various/drinks.csv', nrows=5).drop(columns=['spirit_servings','total_litres_of_pure_alcohol'])
drinks
```

```python
# Create a 3D scatter plot of total liters of pure alcohol consumed per capita
# by continent and country
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')


# continent_map = {s: i for i, s in enumerate(set(drinks.continent))}
# numeric_strings = [string_map[s] for s in drinks.continent]

continent_map = {s: i for i, s in enumerate(set(drinks.continent))}
country_map = {s: i for i, s in enumerate(set(drinks.country))}

x = drinks.continent.map(continent_map)
y = drinks.beer_servings
z = drinks.wine_servings

print(continent_map)
print(x)
print(y)

xticks = continent_map.values()
xtickslabels = continent_map.keys()

ax.set(xticks=list(xticks), xticklabels=list(xtickslabels))

ax.scatter(x,y,z)
ax.set_xlabel('Continent')
ax.set_ylabel('Beer Servings')
ax.set_zlabel('Wine Servings')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Generate some random data
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)  # Color values based on random data

# Create a figure and axis
fig, ax = plt.subplots()

# Define colormap
cmap = matplotlib.colormaps['viridis']  # Choose colormap (viridis in this case)

# Plotting scatter plot with colormap
scatter = ax.scatter(x, y, c=colors, cmap=cmap)

# Adding colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Color Intensity')

# Customizing plot labels and appearance
ax.set_title('Scatter Plot with Colormap')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.grid(True)
plt.tight_layout()
plt.show()
```
