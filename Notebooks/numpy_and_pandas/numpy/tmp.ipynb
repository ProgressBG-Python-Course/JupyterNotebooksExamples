```python
import numpy as np

# Create the indices for rows and columns
rows = [0, -1]
cols = [0, -1]

# Use np.ix_ to create the meshgrid
meshgrid = np.ix_(rows, cols)

# Print the generated meshgrid
print("Meshgrid generated by np.ix_([0, -1], [0, -1]):")
print("Row indices:\n", meshgrid[0])
print("Column indices:\n", meshgrid[1])

# To visualize how it indexes into the original array, create the array
arr = np.arange(1, 10).reshape(3, 3)
print("\nOriginal array:\n", arr)

# Use the meshgrid to extract the corner elements
corner_elements = arr[meshgrid]
print("Corner elements arranged in 2x2 array:\n", corner_elements)
```
